```python
winduv_US = u10_us
```

```python
winduv_US['v10'] = v10_us["v10"]
```

```python
winduv_US["wind.speed"] = numpy.sqrt(winduv_US["u10"] ** 2 + winduv_US["v10"] ** 2)
```

```python
winduv_US['blh'] = blh_us['blh']
```

```python
del winduv_US['v10']
del winduv_US['u10']
```

```python
winduv_US
```

```python
winduv_US.to_netcdf("US_ERA_BLH_wind_2011-2020.nc")
```

```python
winduv_US_df = winduv_US.reset_coords().to_dataframe().reset_index()
```
